# neuroim Tutorials

Welcome to the neuroim tutorial notebooks! These interactive tutorials will guide you through the main features of the package.

## Available Tutorials

- [Image Volumes](image_volumes.ipynb)
- [Neuro Vectors](neuro_vectors.ipynb)
- [Pipelines](pipelines.ipynb)
- [Regions Of Interest](regions_of_interest.ipynb)

## Getting Started

Before running these tutorials, make sure you have neuroim installed:

```bash
pip install neuroim
```

Each notebook contains executable code examples that demonstrate key features of the package. The notebooks are designed to be run in order, but can also be used as standalone references.

### Note on Example Data

These tutorials use synthetic data for demonstration. In practice, you would load real neuroimaging files using functions like `read_vol()` and `read_vec()`.
