# Working with Image Volumes

This tutorial covers the basics of working with 3D neuroimaging volumes in neuroimpy.

## Reading a NIFTI formatted image volume

The primary way to read a volumetric image file is to use `read_vol()`

```python
# Import required libraries
import neuroimpy as pn
import numpy as np

# Create example data for tutorial
# (In practice, you would use real neuroimaging files)

# Create example volume
space_3d = pn.NeuroSpace(dim=(64, 64, 25), spacing=(3.5, 3.5, 5.0), 
                         origin=(-110.5, -88.9342, -42.75))
example_data = np.random.randn(64, 64, 25)
example_data = (example_data - example_data.min()) / (example_data.max() - example_data.min())

# Create volume
vol = pn.DenseNeuroVol(example_data, space_3d)
```

```python
# Note: In practice, you would read a NIFTI file like this:
# vol = pn.read_vol("path/to/image.nii.gz")
```

Working with image volumes

```python
print(vol)
# DenseNeuroVol
#   Type      : DenseNeuroVol
#   Dimension : 64 X 64 X 25
#   Spacing   : 3.5 X 3.5 X 5.0
#   Origin    : -110.5, -88.9342, -42.75
#   Range     : [0.000, 1.000]
```

`read_vol()` returns a `NeuroVol` object which extends a numpy array and has 3 dimensions (x, y, z)

```python
print(type(vol))
print(isinstance(vol.data, np.ndarray))
print(vol.shape)
# Access individual voxels
print(vol[0, 0, 0])
# 0.0
print(vol[63, 63, 24])
# 0.0
```

Arithmetic operations

```python
# Add two volumes
vol2 = vol + vol
assert np.sum(vol2.data) == 2 * np.sum(vol.data)
# Subtract and multiply
vol3 = vol2 - 2 * vol
assert np.all(vol3.data == 0)
# Element-wise operations
vol4 = vol * 2.5
vol5 = vol + 10
vol6 = vol ** 2
```

Converting to logical (binary) volumes

```python
# Convert to logical volume
vol_binary = vol.as_logical()
print(type(vol_binary))
print(vol_binary[0, 0, 0])
```

Creating volumes from arrays

```python
# Create a volume from a numpy array
x = np.zeros((64, 64, 64))
```

Now we create a `NeuroSpace` instance that describes the geometry of the image including its dimensions and voxel spacing

```python
# Create a NeuroSpace instance that describes the geometry of the image
bspace = pn.NeuroSpace(dim=[64, 64, 64], spacing=[1, 1, 1])
vol = pn.DenseNeuroVol(x, bspace)
print(vol)
```

Working with existing spaces

```python
# Create a new volume with the same space as an existing one
vol2 = pn.DenseNeuroVol((vol.data + 1) * 25, vol.space)
print(f"Max value: {np.max(vol2.data)}")
print(f"Volume space: {vol2.space}")
```

Writing a NIFTI formatted image volume

```python
# Write NIFTI formatted image volume
# Note: In practice, you would specify actual file paths
# pn.write_vol(vol2, "output.nii")
# pn.write_vol(vol2, "output.nii.gz")
print("In practice, use pn.write_vol() to save volumes")
```

Sparse volumes

```python
# Create a sparse volume with only a few non-zero values
data = np.zeros((64, 64, 64))
data[10:20, 10:20, 10:20] = 1.0

# Convert to sparse
dense_vol = pn.DenseNeuroVol(data, bspace)
sparse_vol = dense_vol.as_sparse()
print(f"Volume type: {type(sparse_vol)}")

# Sparse volumes behave like dense volumes
print(f"Value at [15, 15, 15]: {sparse_vol[15, 15, 15]}")
print(f"Value at [0, 0, 0]: {sparse_vol[0, 0, 0]}")
```

Volume statistics

```python
# Create a volume with random data
data = np.random.randn(64, 64, 64)
vol = pn.DenseNeuroVol(data, bspace)

# Get statistics
print(f"Mean: {vol.mean():.3f}")
print(f"SD: {vol.std():.3f}")
print(f"Min: {vol.min():.3f}")
print(f"Max: {vol.max():.3f}")

# Find location of minimum and maximum
min_idx = vol.which_min()
max_idx = vol.which_max()

# Convert indices to coordinates
min_coords = vol.space.index_to_grid(min_idx)
max_coords = vol.space.index_to_grid(max_idx)
print(f"Min at grid coordinates: {min_coords[0]}")
print(f"Max at grid coordinates: {max_coords[0]}")
```
